### Rata rata nilai setiap semester

```python
mean_semester = pd.DataFrame(columns=['semester', 'mean'])

for index, item in enumerate(df.columns):
    if index == 0: continue
    semester = f"{item[0]} {item[1]}"
    min_grade = df[(item[0], item[1])].mean()
    mean_semester.loc[index-1] = [semester, min_grade]

mean_semester
```

### min max nilai setiap semester
1. create a dictionary to map year of study to semester
2. create a new dataframe to store the min and max grade for each semester
3. loop through the year_to_semester dictionary

```python
minmax_grade_semester = pd.DataFrame(columns=['semester', 'min_grade', 'max_grade'])

for index, item in enumerate(df.columns):
    if index == 0: continue
    semester = f"{item[0]} {item[1]}"
    min_grade = df[(item[0], item[1])].min()
    max_grade = df[(item[0], item[1])].max()
    minmax_grade_semester.loc[index-1] = [semester, min_grade, max_grade]

minmax_grade_semester
```

```python
highest_grades = pd.DataFrame(columns=['semester', 'nama', 'nilai'])

for index,col in enumerate(df.columns[1:]):
    semester = col[0]
    tahun = col[1]
    nama = df.loc[df[col].idxmax(), ('Unnamed: 0_level_0', 'Nama')]
    nilai = df[col].max()

    highest_grades.loc[index] = [f"{semester} {tahun}", nama,nilai]

highest_grades
```

```python
lowest_grade = pd.DataFrame(columns=['semester', 'nama', 'nilai'])

for index,col in enumerate(df.columns[1:]):
    semester = col[0]
    tahun = col[1]
    nama = df.loc[df[col].idxmin(), ('Unnamed: 0_level_0', 'Nama')]
    nilai = df[col].min()

    lowest_grade.loc[index] = [f"{semester} {tahun}", nama,nilai]

lowest_grade
```

### aplikasi abcd terhadap nilai int

```python
pass_persemester = pd.DataFrame(columns=['semester', 'A', 'B', 'C', 'D', 'E'])

for index, item in enumerate(df.columns):
    if index == 0: continue
    semester = f"{item[0]} {item[1]}"
    a = df[(df[(item[0], item[1])] >= 90) & (df[(item[0], item[1])] <= 100)][(item[0], item[1])].count()
    b = df[(df[(item[0], item[1])] >= 80) & (df[(item[0], item[1])] < 90)][(item[0], item[1])].count()
    c = df[(df[(item[0], item[1])] >= 70) & (df[(item[0], item[1])] < 80)][(item[0], item[1])].count()
    d = df[(df[(item[0], item[1])] >= 60) & (df[(item[0], item[1])] < 70)][(item[0], item[1])].count()
    e = df[df[(item[0], item[1])] < 60][(item[0], item[1])].count()
    
    pass_persemester.loc[index-1] = [semester, a, b, c, d, e]

pass_persemester
```

### nilai naik setiap semester
siswa ab21 nilainya naik atau tetap dari semester sebelumnya

```python
increase = pd.DataFrame(columns=['nama', 'naik', 'tidak'])


for indexx, nama in enumerate(df[("Unnamed: 0_level_0", "Nama")]):
    
    curr = 0
    naik = None
    tidak = None
    for index, item in enumerate(df.columns):
        if index == 0: continue

        nilai_sekarang = df[df[("Unnamed: 0_level_0", "Nama")] == nama][(item[0], item[1])]
        if (nilai_sekarang >= curr).any():
            curr = nilai_sekarang
            naik = True
            tidak = False
        else: 
            naik = False
            tidak = True
            break

    increase.loc[indexx] = [nama, naik, tidak]

increase[increase["naik"] == True]
```
